```python
import tensorflow as tf
```

```python
from dynamic_convolution_layer import DynamiConvolutionLayer
```

```python
inp=tf.random.uniform(shape=(10,256,256,3),minval=0.0,maxval=1.0)
```

```python
inp.shape
```

```python
batch_size=10
classes_config={
    0:3,
    1:4,
    2:2
}
```

```python
conv_layer_config=[
    {
        'filters':16,
        'kernel_size':(3,3),
        'padding':"same",
        'activation':"relu",
        'kernel_initializer':"he_uniform"
    },
     {
        'filters':32,
        'kernel_size':(5,5),
        'padding':"same",
        'activation':"relu",
        'kernel_initializer':"he_uniform"
    }
    
]
```

```python
dense_layer_config=[
    {
        'units':256,
        'activation':"relu",
        'kernel_initializer':"he_uniform"
    },
    {
        'units':32,
        'activation':"relu",
        'kernel_initializer':"he_uniform"
    }
]
```

```python
last_layer_config={
    'activation':"softmax",
    'kernel_initializer':"glorot_uniform"
}
```

```python
pre_dcn_layers_config={
    "ConvolutionLayers":conv_layer_config,
    "DenseLayers":dense_layer_config,
    "LastLayer":last_layer_config
}
```

```python
pred1,pred2=DynamiConvolutionLayer(
    batch_size=batch_size,
    classes_config=classes_config,
    pre_dcn_layers_config=pre_dcn_layers_config
)(inp)
```
